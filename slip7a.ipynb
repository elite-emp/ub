```python


def print_board(board):
    for row in board:
        print(" ".join(row))
    print()

def is_winner(board, player):
    for i in range(3):
        if all([board[i][j] == player for j in range(3)]) or all([board[j][i] == player for j in range(3)]):
            return True
    if board[0][0] == board[1][1] == board[2][2] == player or board[0][2] == board[1][1] == board[2][0] == player:
        return True
    return False

def game_over(board):
    return is_winner(board, 'X') or is_winner(board, 'O') or all([board[i][j] != ' ' for i in range(3) for j in range(3)])

def minimax(board, depth, alpha, beta, is_maximizing):
    if is_winner(board, 'X'):
        return -10
    elif is_winner(board, 'O'):
        return 10
    elif game_over(board):
        return 0
    
    if is_maximizing:
        max_eval = float('-inf')
        for i in range(3):
            for j in range(3):
                if board[i][j] == ' ':
                    board[i][j] = 'O'
                    eval = minimax(board, depth + 1, alpha, beta, False)
                    board[i][j] = ' '
                    max_eval = max(max_eval, eval)
                    alpha = max(alpha, eval)
                    if beta <= alpha:
                        break
        return max_eval
    else:
        min_eval = float('inf')
        for i in range(3):
            for j in range(3):
                if board[i][j] == ' ':
                    board[i][j] = 'X'
                    eval = minimax(board, depth + 1, alpha, beta, True)
                    board[i][j] = ' '
                    min_eval = min(min_eval, eval)
                    beta = min(beta, eval)
                    if beta <= alpha:
                        break
        return min_eval

def find_best_move(board):
    best_eval = float('-inf')
    best_move = (-1, -1)
    
    for i in range(3):
        for j in range(3):
            if board[i][j] == ' ':
                board[i][j] = 'O'
                eval = minimax(board, 0, float('-inf'), float('inf'), False)
                board[i][j] = ' '
                if eval > best_eval:
                    best_eval = eval
                    best_move = (i, j)
    
    return best_move

def play_game():
    board = [[' ' for _ in range(3)] for _ in range(3)]
    print("Initial Board:")
    print_board(board)
    
    while not game_over(board):
        player_move = tuple(map(int, input("Enter your move (row, col): ").split()))
        if board[player_move[0]][player_move[1]] != ' ':
            print("Invalid move. Try again.")
            continue
        board[player_move[0]][player_move[1]] = 'X'
        if game_over(board):
            break
        print("After your move:")
        print_board(board)
        
        ai_move = find_best_move(board)
        board[ai_move[0]][ai_move[1]] = 'O'
        print("After AI's move:")
        print_board(board)
    
    if is_winner(board, 'X'):
        print("Congratulations! You win!")
    elif is_winner(board, 'O'):
        print("AI wins!")
    else:
        print("It's a draw!")

play_game()
```

```python
def check_winner(board):
    # Check rows, columns, and diagonals for a winner
    for i in range(3):
        if board[i][0] == board[i][1] == board[i][2] != ' ':
            return board[i][0]  # Winner in row
        if board[0][i] == board[1][i] == board[2][i] != ' ':
            return board[0][i]  # Winner in column

    if board[0][0] == board[1][1] == board[2][2] != ' ':
        return board[0][0]  # Winner in diagonal
    if board[0][2] == board[1][1] == board[2][0] != ' ':
        return board[0][2]  # Winner in diagonal

    return None  # No winner

def is_board_full(board):
    # Check if the board is full
    for row in board:
        for cell in row:
            if cell == ' ':
                return False  # Board is not full
    return True  # Board is full

def solve_tic_tac_toe(board):
    winner = check_winner(board)
    if winner:
        return f"The winner is {winner}!"
    
    if is_board_full(board):
        return "It's a draw!"

    return "The game is ongoing."

# Example Tic-Tac-Toe board (empty spaces represented by ' ')
example_board = [
    ['X', 'O', 'X'],
    [' ', 'O', ' '],
    ['X', ' ', 'O']
]

# Calling the function to solve the Tic-Tac-Toe problem
solution = solve_tic_tac_toe(example_board)
print(solution)
```
